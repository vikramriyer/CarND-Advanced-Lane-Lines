```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
%matplotlib inline
```

```python
# read using mpimg
img = mpimg.imread('curved-lane.jpg')

# convert to gray
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# show
plt.imshow(gray);
```

```python
# use sobel filter and get result as img*sobel (convolution between img and sobel filter)
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0)

# take abs values, may introduce some floats probably
abs_sobelx = np.absolute(sobelx)

# scale down
scaled_sobel = np.uint8(255*abs_sobelx/np.max(abs_sobelx))
```

```python
# set threasholds (the values between min and max will be set as 1)
thresh_min = 20
thresh_max = 100

# create identical array with all zeros
sxbinary = np.zeros_like(scaled_sobel)

# set values based on thresholds
sxbinary[(scaled_sobel >= thresh_min) & (scaled_sobel <= thresh_max)] = 1
plt.imshow(sxbinary, cmap='gray');
```

```python
# use sobel filter and get result as img*sobel (convolution between img and sobel filter)
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1)

# take abs values, may introduce some floats probably
abs_sobely = np.absolute(sobely)

# scale down
scaled_sobely = np.uint8(255*abs_sobely/np.max(abs_sobely))
```

```python
# set threasholds (the values between min and max will be set as 1)
thresh_min = 20
thresh_max = 100

# create identical array with all zeros
sybinary = np.zeros_like(scaled_sobely)

# set values based on thresholds
sybinary[(scaled_sobely >= thresh_min) & (scaled_sobely <= thresh_max)] = 1
plt.imshow(sybinary, cmap='gray');
```
