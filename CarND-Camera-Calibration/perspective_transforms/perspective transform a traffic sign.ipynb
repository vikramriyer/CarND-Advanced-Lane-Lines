```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
img = cv2.imread('funnyroad_2.jpg')
```

```python
plt.imshow(img);
plt.plot(860,330,'.',color='red', marker='o');
plt.plot(860,700,'.',color='red', marker='o');
plt.plot(1080,420,'.',color='red', marker='o');
plt.plot(1090,750,'.',color='red', marker='o');
```

```python
def warp(img):
    
    img_size = (img.shape[1], img.shape[0])
    
    # set the source and destination points
    src = np.float32([[(860,330), (860,750), (1300,420), (1300,750)]])
    dst = np.float32([[(780,430), (780,730), (1200,370), (1200,700)]])
    
    # get the transformed image
    M = cv2.getPerspectiveTransform(src, dst)
    
    # keep the inverse transformed image so that we can get the original back
    MInv = cv2.getPerspectiveTransform(dst, src)
    
    # warp the image
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    
    # return the warped image
    return warped
```

```python
warped_img = warp(img)

# visualizations
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(img);
ax2.imshow(warped_img);
```
